## **Normal (Gaussian) distributions**
The Gaussian distribution is arguably the most useful probability distributions out there. The reason is that many random processes are described well by a Gaussian distribution when the number of samples is large (for more on why, look up the central limit theorem).

When data have a Gaussian distribution, there are some extra mathematical tricks that we can use to analyze those data. One of the reasons for this is that (if we have sampled the data sufficiently) we can describe the data with two variables: the mean ($\mu$) and the variance ($\sigma^2$).

This is the equation for a Gaussian distribution:

$f(x) = \frac{1}{\sigma \sqrt{2 \pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

$\mu$ is the mean of the distribution

$\sigma$ is the standard deviation of the distribution (square root of the variance)

It's alright if equations aren't your thing. The thing to remember is that you have some distribution of your data and that there is a mean/average ($\mu$) and another parameter describing how spread out the distribution is---the standard deviation ($\sigma$). The following code allows us to play with those two parameters and see how it affects the Gaussian distribution. Go ahead and change 'mu' and 'sigma' and then push the "play" button to see how the distribution changes.

```
# Numpy module
import numpy as np

# Seaborn for visualizing distributions
import seaborn as sns

numSamples = 10000;
numBins = 25;

# The distribution mean
mu = 1.0

# The standard deviation (sigma).
sigma = 0.3

# 'randn' just generates a bunch of random numbers that have a Gaussian distribution
y = sigma * np.random.randn(numSamples) + mu


ax = sns.histplot(data=y, stat="probability", bins=numBins, kde=True, element='poly')
ax.set(xlabel='Gaussian distribution')
```

Go ahead and play around with the parameters for the Gaussian distribution. What you'll notice as you change the parameters, is that the values along the x-axis change, but the general, bell-curve, shape stays the same.

Gaussian functions will keep coming up, not just in understanding neuroscience concepts, but also in your data analysis. For example, many of the tools that simplify statistical calculations (such as t-tests) assume that your data conforms to a Gaussian distribution. If you're not certain that this is the case, then you need to use other methods.



## **Poisson distributions**

A Poisson distribution is a very important type of distribution in neuroscience. This type of distribution is discrete, meaning that every data entry is an integer value such as 0, 1, 2... (not 1.3439034).

Why is the Poisson distribution so useful? Well, it turns out that there are many physical processes that are well described with Poisson statistics. Here are a few examples that are relevant to you and me:

1) In physics, the frequency of photon arrivals.

2) Neural spike counts (not exactly due to the refractory period)...

3) Frequency of vesicle release at synapses

One of the key properties of the Poisson distribution is how many variables are required to describe them. You'll recall from our discussion of Gaussian distributions that we need two numbers to describe them: 1) the mean ($\mu$) tells you where the distribution is centered and 2) the variance ($\sigma^2$) tells you how spread out the distribution is. Well, a Poisson distribution is even easier, you just need one number to describe it because the mean and variance are equal.

The standard deviation ($\sigma$) is the square root of the variance and, since the mean and variance are equal, it is also the square root of the mean ($\mu$).

$\sigma^2 = \mu$

$\sigma = \sqrt{\mu}$


Let's look at concrete example. Photon arrivals at a detector obey Poisson statistics so let's simulate a distribution of photon counts based on a mean values.

```
# Numpy module
import numpy as np

import math

# Seaborn for visualizing distributions
import seaborn as sns

numSamples = 10000;

# Define the mean/variance of the distribution
mu = 3.0

# This is a way of generating a Poisson distribution.
y = np.random.poisson(mu, size=numSamples)

ax = sns.histplot(data=y, stat="probability", discrete=True)
ax.set(xlabel='photon count per bin')
```

## **Uniform distributions**

Data can also be distributed uniformly: a case in which all of the possible outcomes are equally likely. If we think of this in the context of the photon count example above, having a count of {0,1,2,...} would occur with equal probability. 

Let's look at some code for generating a visualizing data with a uniform distribution.



```
# Numpy module
import numpy as np

# Seaborn for visualizing distributions
import seaborn as sns

numSamples = 10000;
numBins = 25;

y = np.random.rand(numSamples)


ax = sns.histplot(data=y, stat="probability", bins=numBins, kde=False, element='poly')
ax.set(xlabel='Uniform distribution')
```

## **Binomial distributions**

Binomial distributions are an important type of discrete distribution (as opposed to a continuous distribution, such as a Gaussian distribution).

$f(k; p) = \big( \frac{n!}{k!(n-k)!} \big) p^k (1 - p)^{n-k}$ for k $\in$ {0, 1, ..., n}

Let's look at an example of how this distribution might be used. Let's say I give you a coin and ask you to flip the coin five times (that's your $n$ variable). If you know it's a fair coin, what are the chances that you end up with zero heads or one heads, etc? That's your $k$ variable! 


```
# Numpy module
import numpy as np

# Seaborn for visualizing distributions
import seaborn as sns

numSamples = 10000;

n = 5

# This is the probability that a value of '1' is observed. Go ahead and play 
# with this value and see how the distribution changes. It needs to be a value 
# between 0-1. For a fair coin, the value should be 0.5.
p = 0.5

rng = np.random.default_rng()
y = rng.binomial(n, p, numSamples)

ax = sns.histplot(data=y, stat="probability", discrete=True)
ax.set(xlabel='Binomial distribution')
```

## **Bernoulli distribution**

The Bernoulli distribution is a type of binomial distribution. Data with this distribution have values that are either 0 or 1.

$f(k; p) = p^k (1 - p)^{1-k}$ for k $\in$ {0, 1}



```
# Numpy module
import numpy as np

# Seaborn for visualizing distributions
import seaborn as sns

numSamples = 10000;

n = 1

# This is the probability that a value of '1' is observed. Go ahead and play 
# with this value and see how the distribution changes. It needs to be a value 
# between 0-1
p = 0.1

rng = np.random.default_rng()
y = rng.binomial(n, p, numSamples)

ax = sns.histplot(data=y, stat="probability")
ax.set(xlabel='Bernoulli distribution')
```
